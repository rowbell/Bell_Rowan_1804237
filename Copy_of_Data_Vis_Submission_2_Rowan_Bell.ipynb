Combining Datasets

```python
#Combining Datasets

#Renaming Columns
data1 = pd.read_csv('dems.csv').rename(columns={'Gaming penetration in the United Kingdom (UK) from 2013 to 2021, by age group and gender' : 'type',
                                                'Unnamed: 1' : '2013',
'Unnamed: 2' : '2014', 
'Unnamed: 3' :'2015',
'Unnamed: 4' : '2016',
  'Unnamed: 5' : '2017', 
  'Unnamed: 6' : '2018',
   'Unnamed: 7' : '2019',
    'Unnamed: 8' : '2020/21'})

#Dropping Unwanted Columns
data1 = data1.drop(0)
data1 = data1.drop(1)
data1 = data1.replace("-", "0%")
data1 = data1.rename(lambda x: data1['type'][x])
data1 = data1.drop('type', axis=1)

data1


def sanitizeDate(date):
    group = re.search("([0-9]{4})", date)
    return int(group.group()) if not group == None else None

#Renaming Columns
data = pd.read_csv('data.csv').rename(columns={'Initial  release date' : 'Release Date'})

data['Release Date'] = data['Release Date'].apply(lambda rawDate: sanitizeDate(rawDate))

#Filtering Data 

data = data[data['Release Date']>= 2013]


for date in data['Release Date'].tolist():
    print(int(date))
    


data = data.drop('Unnamed: 0', axis=1)

#Appending data from dataset 1 to dataset 2
males = []
females = []
for date in data['Release Date']:
    males.append(data1[str(int(date))]['Male'])
    females.append(data1[str(int(date))]['Female'])


males
data['Males'] = males
data['Females'] = females

data

```
